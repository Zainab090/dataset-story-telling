```python
# 📘 Housing Prices Storytelling Notebook (Beginner Friendly – No ML)

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
df = pd.read_csv("/kaggle/input/housing-prices-dataset/Housing.csv")

# Step 3: Quick Look at Data
print("First 5 rows:")
print(df.head())
print("\nShape of dataset:", df.shape)

# Step 4: Dataset Info
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())

# Step 5: Basic Stats
print("\nBasic Statistics:")
print(df.describe())

# Step 6: Storytelling with Visuals

# 6.1 Price Distribution
plt.figure(figsize=(8,6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Act I: How House Prices are Spread")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# 6.2 Area vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['area'], y=df['price'])
plt.title("Act II: Bigger Houses Usually Cost More")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()

# 6.3 Bedrooms vs Price
plt.figure(figsize=(8,6))
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title("Act III: More Bedrooms = Higher Price?")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")
plt.show()

# 6.4 Furnishing Status vs Price
plt.figure(figsize=(8,6))
sns.barplot(x="furnishingstatus", y="price", data=df, estimator="mean", ci=None)
plt.title("Act IV: Furnished Houses Cost More on Average")
plt.xlabel("Furnishing Status")
plt.ylabel("Average Price")
plt.show()

# Step 7: Story Summary
print("\n📖 Housing Prices Storyline:")
print("Act I: Prices vary widely, with some very expensive houses.")
print("Act II: Larger area usually means higher price.")
print("Act III: Houses with more bedrooms tend to be costlier.")
print("Act IV: Fully furnished houses are generally the most expensive.")
```
